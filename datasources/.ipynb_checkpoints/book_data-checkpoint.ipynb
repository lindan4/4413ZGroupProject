```python
import pandas
pandas.__version__
```

```python
data = pandas.read_csv('book_data.csv')
data
```

```python
data.shape
```

```python
data_updated = data.dropna()
data_updated.shape
```

```python
if 'book_pages' in data_updated:
    data_updated['book_pages'].replace('\s*pages$', '', regex=True, inplace=True)
if 'book_review_count':
    data_updated.drop(columns=['book_review_count'], inplace=True)
```

```python
data_completed = data_updated.rename(columns={ 
    "book_authors": "authors", 
    "book_desc": "description", 
    "book_edition": "edition",
    "book_format": "format",
    "book_isbn": "isbn",
    "book_pages": "pages",
    "book_rating": "rating",
    "book_rating_count": "rating_count",
    "book_title": "title"
})
data_completed
```

```python
data_completed['bid'] = ["b{}".format('{}'.format(x).zfill(4)) for x in range(1, len(data_completed) + 1)]
data_completed
```

```python
data_completed.to_csv("book_data_updated.csv", index=False)
```
