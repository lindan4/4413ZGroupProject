```python
import pandas
pandas.__version__
```

```python
book_data = pandas.read_csv('book_data.csv')
book_data.head()
```

```python
book_data.shape
```

```python
book_data.dropna(inplace=True)
book_data.shape
```

```python
if 'book_pages' in book_data:
    book_data['book_pages'].replace('\s*pages$', '', regex=True, inplace=True)
if 'book_review_count' in book_data:
    book_data.drop(columns=['book_review_count'], inplace=True)
```

```python
book_data = book_data.rename(columns={ 
    "book_authors": "authors", 
    "book_desc": "description", 
    "book_edition": "edition",
    "book_format": "format",
    "book_isbn": "isbn",
    "book_pages": "pages",
    "book_rating": "rating",
    "book_rating_count": "rating_count",
    "book_title": "title",
    "genres": "category"
})
book_data.head()
```

```python
import random

book_data['bid'] = ["b{}".format('{}'.format(x).zfill(4)) for x in range(1, len(book_data) + 1)]
book_data['price'] = [round(random.uniform(6.0, 50.0), 2) for _ in range(1, len(book_data) + 1)]
book_data.head()
```

```python
import collections
authors_map = {}

for authorList in book_data['authors']:
    authors = authorList.split("|")
    for author in authors:
        if author not in authors_map:
            authors_map[author] = len(authors_map)
len(authors_map)
```

```python
data =  []
for (author, author_id) in authors_map.items():
    data.append([author_id, author])
author_data = pandas.DataFrame(data, columns = ['author_id', 'author'])
author_data
```

```python
author_data.to_csv("author_data.csv", index=False)
```

```python
data = []
for (bid, authorList) in zip(book_data['bid'], book_data['authors']):
    authors = authorList.split("|")
    for author in authors:
        data.append([bid, authors_map[author]])

book_authors = pandas.DataFrame(data, columns = ['bid', 'author_id'])
book_authors["id"] = book_authors.index + 1
book_authors.head()
```

```python
book_authors.to_csv("book_authors.csv", index=False)
```

```python
category_map = {}

for categoryList in book_data['category']:
    categories = categoryList.split("|")
    for category in categories:
        if category not in category_map:
            category_map[category] = 0
        category_map[category] += 1
len(category_map)
```

```python
category_counter = {k: v for k, v in sorted(category_map.items(), key=lambda item: item[1])}
category_counter
```

```python
# use 75 as threshold
```

```python
data =  []
category_map = {}
for (category_id, (category, category_count)) in enumerate(category_counter.items()):
    if (category_count >= 75):
        data.append([category_id, category])
        category_map[category] = category_id
    
category_data = pandas.DataFrame(data, columns = ['category_id', 'category'])
category_data.head()
```

```python
category_data.to_csv("category_data.csv", index=False)
```

```python
data = []
for (bid, authorList) in zip(book_data['bid'], book_data['category']):
    categories = categoryList.split("|")
    for category in categories:
        if category in category_map:
            data.append([bid, category_map[category]])

book_categories = pandas.DataFrame(data, columns = ['bid', 'category_id'])
book_categories["id"] = book_categories.index + 1
book_categories.head()
```

```python
book_categories.to_csv("book_categories.csv", index=False)
```
